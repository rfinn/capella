```python
from matplotlib import pyplot as plt
import numpy as np
#from numpy.lib.recfunctions import append_fields
from astropy.io import fits
import glob
%matplotlib inline
```

```python
class Cluster:
    def __init__(self,prefix):
        self.prefix = prefix
        # read in NSA
        nsafile = prefix+'_NSA.fits'
        self.nsa = fits.getdata(nsafile)
        # read in AGC
        agcfile= prefix+'_matchAGCtoNSA.fits'
        self.agc = fits.getdata(agcfile)
    def plotpositions(self):
        plt.figure()
        plt.plot(self.nsa.RA,self.nsa.DEC,'k.')
    def plotcolormag(self):
        plt.figure()
        self.NUVr=self.nsa.ABSMAG[:,1] - self.nsa.ABSMAG[:,4]
        plt.plot(self.nsa.ABSMAG[:,4],self.NUVr,'bo')
        
        
```

```python
clusters=['Abell2063']
a2063 = Cluster(clusters[0])
```
